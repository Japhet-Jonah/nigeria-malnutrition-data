```python
columns_to_round = ['Severe Wasting', 'Wasting', 'Overweight', 'Stunting', 'Underweight']
print(apps_data.columns)
apps_data[columns_to_round] = apps_data[columns_to_round].round(1)
display(apps_data.head())
```

```python
apps_data = apps_data.rename(columns={"U5 Population ('000s)": "U5 Population"})
```

```python
income_mapping = {
    0: 'Low Income',
    1: 'Lower-middle Income',
    2: 'Upper-Middle Income',
    3: 'High Income'
}
apps_data['Income Classification'] = apps_data['Income Classification'].map(income_mapping)
display(apps_data.head())
```

```python
apps_data.tail(10)
```

```python
malnutrition_indicators = ['Severe Wasting', 'Wasting', 'Overweight', 'Stunting', 'Underweight']

for indicator in malnutrition_indicators:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Income Classification', y=indicator, data=apps_data, ci=None)
    plt.title(f'Average {indicator} by Income Classification')
    plt.ylabel(indicator)
    plt.xlabel('Income Classification')
    plt.show()
```

```python
malnutrition_indicators = ['Severe Wasting', 'Wasting', 'Overweight', 'Stunting', 'Underweight']

for indicator in malnutrition_indicators:
    fig = px.choropleth(apps_data,
                        locations="Country",
                        locationmode='country names',
                        color=indicator,
                        hover_name="Country",
                        title=f'World Map of {indicator}',
                        color_continuous_scale=px.colors.sequential.Plasma)
    fig.show()
```

```python
malnutrition_indicators = ['Severe Wasting', 'Wasting', 'Overweight', 'Stunting', 'Underweight']
correlation_matrix = apps_data[malnutrition_indicators].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Malnutrition Indicators')
plt.show()
```

```python
ranked_by_severe_wasting = apps_data.sort_values(by='Severe Wasting', ascending=False)
display(ranked_by_severe_wasting[['Country', 'Severe Wasting']].head(10))
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Severe Wasting', data=ranked_by_severe_wasting.head(10), palette='viridis')
plt.title('Top 10 Countries by Severe Wasting')
plt.xlabel('Country')
plt.ylabel('Severe Wasting (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
ranked_by_severe_wasting = apps_data.sort_values(by='Stunting', ascending=False)
display(ranked_by_severe_wasting[['Country', 'Stunting']].head(10))
```

Seaborn is a powerful library for creating statistical visualizations. Some of the common types of plots you can generate with Seaborn include:

* **Distribution plots:** Histograms, kernel density estimates (KDE plots), and box plots to visualize the distribution of a single variable.
* **Relational plots:** Scatter plots and line plots to visualize the relationship between two or more variables.
* **Categorical plots:** Bar plots, count plots, box plots, violin plots, and swarm plots to visualize the relationship between a numerical variable and a categorical variable.
* **Regression plots:** Plots to visualize linear relationships and their confidence intervals.
* **Matrix plots:** Heatmaps and clustered heatmaps to visualize relationships between multiple variables or correlations.
* **Multi-plot grids:** Creating grids of plots to visualize relationships across different subsets of your data.
